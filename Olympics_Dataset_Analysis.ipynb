```python
import pandas as pd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('summer.csv')
```

## 1. In how many cities Summer Olympics is held so far?


```python
data = df['City'].unique()
for i in data:
    print(i)
```

## 2. Which sport is having most number of Gold Medals so far? (Top 5)


```python
gold_medal = df[df["Medal"] == "Gold"]         # Selecting only Gold medal from df dataset

data = []

for sport in gold_medal["Sport"].unique():     # Selecting sport for gold medal in gold_medal dataset
    
    data.append([sport, len(gold_medal[gold_medal["Sport"] == sport])])

pd.DataFrame(data, columns = ['sport', 'Golds']).head().sort_values(by = 'Golds', ascending = False).plot(x = 'sport', y = 'Golds', kind = 'bar', figsize = (10,5), color = 'red', xlabel = 'Sport', ylabel = 'Winning Golds')
```

## 3. Which sport is having most number of medals so far? (Top 5)


```python
medal_rank = df.groupby("Sport")["Medal"].apply(lambda x: x.notnull().sum()).reset_index(name = 'Medal')
medal_rank = medal_rank.sort_values(by = 'Medal',ascending = False)
medal = medal_rank.head(5)
sns.catplot(x = "Sport", y = "Medal",kind = "bar", data = medal, aspect = 20/8.27)
```

## 4. Which player has won most number of medals? (Top 5)


```python
data = []

for player in df['Athlete'].unique():     # no of times given player name occurs is the times of when given player wins medal.
    data.append([player , len(df[df['Athlete']  == player])])

pd.DataFrame(data,columns = ['Player','medals']).sort_values(by='medals', ascending=False).head().plot(x = 'Player', y = 'medals', kind = 'bar', figsize = (10,5), xlabel = 'Player', ylabel = 'Overall medals')
```

## 5. Which player has won most number Gold Medals of medals? (Top 5)


```python
gold_medal = df[df["Medal"] == "Gold"]         # Selecting only Gold medal from df dataset

data = []

for Athlete in gold_medal["Athlete"].unique():     # Selecting sport for gold medal in gold_medal dataset
    
    data.append([Athlete, len(gold_medal[gold_medal["Athlete"] == Athlete])])

pd.DataFrame(data, columns = ['Athlete', 'Golds']).head().sort_values(by = 'Golds', ascending = False).plot(x = 'Athlete', y = 'Golds', kind = 'bar', figsize = (10,5), color = 'blue', xlabel = 'player', ylabel = 'Winning Golds')

```

## 6. In which year India won first Gold Medal in Summer Olympics?


```python
data = []
gold_medal = df[df["Medal"] == "Gold"]                                
country = gold_medal[gold_medal["Country"] == "IND"]                   
for year in country["Year"]:                                           
    data.append(year)                                                  
print(data[0])
```

## 7. Which event is most popular in terms on number of players? (Top 5)


```python
data = []

for Event in df['Event'].unique():
    data.append([Event, len(df[df['Event'] == Event])])
    
pd.DataFrame(data,columns = ['Event','Players']).sort_values(by='Players', ascending=False).head().plot(x = 'Event', y = 'Players', kind = 'bar', figsize = (10,5), xlabel = 'Event', ylabel = 'Players')
```
